## 1. 구단별 홈, 어웨이 승률 구하기

홈승률 =  홈에서 승/(홈에서 승 + 홈에서 패)                     
어웨이 승률 = 원정 승/(원정 승 + 원정 패)

```python
def home_away_winrate(df, season):
    new = {}
    team_list = ['WO','OB','NC','SK','LG','KT','LT','HT','SS','HH']
    home_rate_list=[]
    away_rate_list=[]
    df = df[df['시즌'] == season]
    for team in team_list:
        df_team = df[df['팀코드'] == team]
        home_win=0
        home_lose=0
        home_draw = 0
        away_win=0
        away_lose=0
        away_draw = 0
        for i, row in df_team.iterrows():
            if df_team.at[i,'초말'] == 'B':
                if df_team.at[i,'승패'] == 'W':
                    home_win+=1
                elif df_team.at[i,'승패'] == 'L':
                    home_lose+=1
                else:
                    home_draw+=1
            elif df_team.at[i,'초말'] == 'T':
                if df_team.at[i,'승패'] == 'W':
                    away_win+=1
                elif df_team.at[i,'승패'] == 'L':
                    away_lose+=1
                else:
                    away_draw+=1
        home_rate = home_win/(home_win+home_lose)
        away_rate = away_win/(away_win+away_lose)
        home_rate_list.append(home_rate)
        away_rate_list.append(away_rate)
    col1_name = '홈승률'+str(season)
    col2_name = '원정승률'+str(season)
    new['구단'] = team_list
    new[col1_name] = home_rate_list
    new[col2_name] = away_rate_list
    df_new = pd.DataFrame(new)
    return df_new
```

## 2. 1을 이용해 구단의 시즌별 승률 구하기

```python
def home_away_winrate_season(df):
    seasons = [2016,2017,2018,2019,2020]
    first = True
    for season in seasons:
        df_season = home_away_winrate(df, season)
        if first:
            ha_winrate = df_season
            first = False
        else:
            ha_winrate=pd.merge(ha_winrate,df_season)
    return ha_winrate
```
